```python
from scipy.stats import linregress
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()

x = np.random.randn(1000)
y = 2 * x + 0.1 * np.random.randn(1000)
gradient, intercept, r, p, std_err = linregress(x, y)

y_pred = gradient * x + intercept
ax.plot(x, y, color='red')
ax.plot(x, y_pred, color='blue')
```

```python
f = lambda x: np.sin(4 * (x - 1/4)) + x + x**20 - 1
x = np.linspace(0, 1, 200)

fig, ax = plt.subplots(figsize=(10, 8))
ax.plot(x, f(x))
ax.axhline(ls='--', c='k')
plt.show()
```

Finding roots:

```python
def bisect(f, a, b, tol=10e-5):
    """
    Implements the bisection root finding algorithm, assuming that f is a real-valued function on [a, b]
    satisfying f(a) < 0 < f(b).
    """
    
    lower, upper = a, b
    while lower - upper > tol:
        middle = 0.5 * (upper + lower)
        # if root b/w lower and middle
        if f(middle) > 0:
            lower, upper = lower, middle
        else:
            lower, upper = middle, upper
    
    return 0.5 * (upper + lower)
```

```python
from scipy.optimize import newton

newton(f, 0.2)
# initial condition of x = 0.2
```

```python
newton(f, 0.7) # converges to local optima
```

uses local slope information => similar to gradient descent.
while unreliable sometimes, much faster than bisection method
