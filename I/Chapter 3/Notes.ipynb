```python
ts_length = 100
values = []

for i in range(ts_length):
    e = np.random.randn()
    values.append(e)
    
plt.plot(values)
plt.show()
```

```python
def generate_data(n, generator_type):
    values = []
    for i in range(n):
        e = generator_type()
        values.append(e)
    return values

data = generate_data(100, np.random.uniform)
plt.plot(data)
plt.show()
```

```python
def generate_data(n, generator_type):
    return [generator_type() for i in range(n)]

data = generate_data(100, np.random.uniform)
plt.plot(data)
plt.show()
```
