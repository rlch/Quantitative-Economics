Exercise 1

```python
def factorial(n):
    result = 1
    for i in range(n):
        result *= (i + 1)
    return result

factorial(6)
```

Exercise 2

```python
from numpy.random import uniform

def binomial_rv(n, p):
    return sum([uniform() < p for i in range(n)])

binomial_rv(10, 0.5)
```

Exercise 3

```python
import numpy as np

n = 1000000

hit = 0
for i in range(n):
    u, v = np.random.uniform(), np.random.uniform()
    d = np.sqrt((u - 0.5)**2 + (v - 0.5)**2)
    if d < 0.5:
        hit += 1
        
area = hit / n
print(area * 4)
```

Exercise 4

```python
from numpy.random import uniform

def realization(n):
    count = 0
    for i in range(n):
        count = count + 1 if uniform() < 0.5 else 0
        if count == 3:
            return 1
    return 0

ev = sum([realization(10) for _ in range(1000)]) / 1000
print(ev)
```

```python
# probably overkill using a function but whatever
def ts(T, α):
    x = [0]
    for i in range(T):
        x.append(α * x[i - 1] + np.random.randn())
    return x
    
for α in [0, 0.8, 0.98]:
    plt.plot(ts(200, α), label=f"α = {α}")
plt.legend()
plt.show()
```
